## What is numpy

Numpy is the core library for scientific and numerical computing in python.

It provides high-performence multidimensional array object and tools for working with arrys.

numpy's main objective to work with multidimentional arrays

It is a table of elements (usually numbers), all of the same type. indexed by a tuple of positive integers.

In numpy, dimensions are called axes.

```python
c = np.arange(100)
print(c.size*c.itemsize)
```

```python
size = 10000
```

```python
# this is the list
l1 = range(size)
l2 = range(size)
# this is the numpy array
a1 = np.arange(size)
a2 = np.arange(size)
```

```python
start = time.time()
result = [(x+y) for x,y in zip(l1,l2)]
print(result)
print("python list took time: ", (time.time()-start)*1000)
```

```python
start = time.time()
result = a1+a2
print("Numpy array took the time: ", (time.time()-start)*1000)
```
