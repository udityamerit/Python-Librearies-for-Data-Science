## 1. What is the Matplotlib 

It is an open source drawing  library which supports rich drawing types.

It is used to plot or draw 2D and 3D graphics for the Visualization of the data.

You can understand your data easily by visualizing it with the help of matplotlib

you can generate plots, histograms, bar charts and many others with just a few lines of code


## 2. Types of plots

plot(x, y): Line plot, connects data points with straight lines.

scatter(x, y): Scatter plot, displays data points as individual markers.

bar(x, height): Vertical bar chart.

barh(y, width): Horizontal bar chart.

stem(x, y): Stem plot, displays data points as vertical lines with markers.

fill_between(x, y1, y2): Fills the area between two curves.

stackplot(x, y): Stacked area chart.

stairs(values): Stairstep plot.

## 6. Line graphs - line types, color and transparency of the graphs.

```python
plt.figure(figsize = (10,4))
plt.subplots(dpi = 150)   # dpi = dots per inch
plt.plot(x,x+1, color = 'r', alpha = 1)
plt.plot(x,x+2, color = 'g', alpha = .8)
plt.plot(x,x+3, color = 'y', alpha = .6)

plt.legend(["y = x+1", "y = x+2", "y = x+3"], loc = 4)
plt.show()
```

```python
plt.figure(figsize = (10,4))
plt.subplots(dpi = 150)   # dpi = dots per inch
plt.plot(x,x**1, color = 'r', lw = .8) # you can use the lw for the linewidth(lw)
plt.plot(x,x**2, color = 'g', linewidth = 1.5)
plt.plot(x,x**3, color = 'y', linewidth = 2.5)

plt.legend(["y = x**1", "y = x**2", "y = x**3"], loc = 2)
plt.show()
```

```python
# supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

plt.figure(figsize = (10,4))
plt.subplots(dpi = 150)   # dpi = dots per inch
plt.plot(x,x+1, color = 'r', lw = 2, ls = 'dashdot') # you can use the ls for the linestyle(ls)
plt.plot(x,x+2, color = 'g', lw= 1.5, linestyle = ':')
plt.plot(x,x+3, color = 'y', lw = 2, ls = '--')
line, = plt.plot(x, x+4, color = 'black', lw = 1.2)
line.set_dashes([5,10,4,14])

plt.legend(["y = x+1", "y = x+2", "y = x+3", "y = x+4"], loc = 4)


plt.show()
```

```python
plt.figure(figsize = (10,4))
plt.subplots(dpi = 150)   # dpi = dots per inch for clear image or plot
plt.plot(x,x+1, color = 'r',  marker = 'o', markersize = 3) # you can use the ls for the linestyle(ls)
plt.plot(x,x+2, color = 'g',  marker = '+')
plt.plot(x,x+3, color = 'y', marker = 's', markerfacecolor = 'red')
line, = plt.plot(x, x+4, color = 'black', marker = '1')
line.set_dashes([5,10,4,14])

plt.legend(["y = x+1", "y = x+2", "y = x+3", "y = x+4"], loc = 4)


plt.show()
```

 ## 7. Canvas grid and axis range 

```python
fig, axes = plt.subplots(1,2, figsize = (10,5), dpi = 200)

axes[0].plot(x, x**2, x, x**3, lw = 2) # here axes[0] denotes the first subplot
axes[0].grid(True)
axes[0].legend(['y = x**2', 'y = x**3'])
axes[1].plot(x, x**2, x, x**3, lw = 2) # here axes[1] denotes the second subplot

axes[1].set_ylim([0,60])
axes[1].set_xlim(2,8)
axes[1].legend(['y = x**2', 'y = x**3'])

'''After apply the set_ylim and set_xlim you can see the plot of subplot 2 the starting points of the plots
and the x and y limit is resetricted based on the given parameter or constrains'''

axes[1].grid(True)
```

## 8. 2D plots - scatter, step, bar, fill_between. 

```python
n = np.array([0,1,2,3,4,5])

fig,  axes = plt.subplots(1, 4, figsize= (16,5))# it is create the 4 different canvas 

axes[0].set_title('Scatterplot', size = 20, color = 'red')
axes[0].scatter(x, x+0.25*np.random.randn(len(x)),color = 'r')

axes[1].set_title('Stepplot', size = 20, color = 'green')
axes[1].step(n, n**2, lw = 3, color = 'g')

axes[2].set_title('Barplot', size = 20, color = 'y')
axes[2].bar(n, n**2, align = 'center', width = 0.7, alpha = 0.8, color = 'y')

axes[3].set_title('Fill Between plot', size = 20, color = 'b')
axes[3].fill_between(x, x**2, x ,x**4, color = 'b', alpha = .9)
```

### Histogram 

```python
n = np.random.randn(10000)
fig, axes = plt.subplots(1,2, figsize = (12,4))

axes[0].set_title("Default histogram", size = 20, color ='b')
axes[0].hist(n)

axes[1].set_title("Cumulative detailed histogram", size = 20, color ='r')
axes[1].hist(n, cumulative = True, bins = 40, color = 'r')
```

### Draw contour image

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm # cm means that colormap
import matplotlib.pyplot as plt
```

```python
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1-Z2)*2
print(X)
print(Y)
print(Z)
```

```python
fig, ax = plt.subplots(figsize = (10,6), dpi = 200)
CS = ax.contour(X,Y,Z)
ax.clabel(CS, inline = 1, fontsize = 10)
ax.set_title("Contour map", size = 20)
```
